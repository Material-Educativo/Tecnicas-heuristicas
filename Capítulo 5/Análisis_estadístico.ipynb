#Análisis estadístico
En este cuaderno de trabajo se incluye el código necesario para realizar un análiss estadistico de las observaciones disponibles.

Las siguientes celdas muestran como obtener la media, mediana, varianza, cuartiles, rango intercuartil y diagramas de caja.

El primerr paso es importar la librería numpy para el cálculo de los estadísticos, y la librería matploblib para realizar los diagrramas de cajas.

```python
import numpy as np
import matplotlib.pyplot as plt
```

Como en este momento no se tienen observaciones reales se generarán dos conjuntos de datos con valores aleatorios.

```python
# Se fija la semilla para reproducibilidad
np.random.seed(42)

# Generamos 30 valores aleatorios entre 0 y 100
# Simulan costos del algoritmo A
datos1 = np.round(np.random.rand(30) * 100, 2)
```

```python
# Se imprimen los datos para poder verlos
print(datos1)
```

```python
# Generamos 30 valores aleatorios entre 150 y 250
# Simulan costos del algoritmo B (peor que A)
datos2 = np.round((np.random.rand(30) + 1.5) * 100, 2)
```

```python
# Nuevamente se imprimen los datos
print(datos2)
```

# Cálculo de estadísticos
En las siguientes celdas se calculará la media, mediana, desviación estádar, cuartiles y rango intercuartil de los datos generados de forma aleatoria.

Ahora se emplea la biblioteca numpy para calcular los estadísticos deseados.

```python
# Calculo de la media
media_datos1 = np.mean(datos1)
media_datos2 = np.mean(datos2)

# Calculo de la mediana
mediana_datos1 = np.median(datos1)
mediana_datos2 = np.median(datos2)

# Calculo de la varianza muestral
varianza_datos1 = np.var(datos1, ddof=1)
varianza_datos2 = np.var(datos2, ddof=1)

# Calculo de los cuartiles
cuartiles_datos1 = np.percentile(datos1, [25, 50, 75])
cuartiles_datos2 = np.percentile(datos2, [25, 50, 75])

# Calculo del rango intercuartil
rango_intercuartil_datos1 = np.diff(cuartiles_datos1)
rango_intercuartil_datos2 = np.diff(cuartiles_datos2)
```

Ahora imprimimos la información.

```python
# Resultados para el primer conjunto de datos
print("="*50)
print("ALGORITMO A (datos1)")
print("="*50)
print(f"Media:                {media_datos1:.2f}")
print(f"Mediana:              {mediana_datos1:.2f}")
print(f"Varianza:             {varianza_datos1:.2f}")
print(f"Desviacion estandar:  {np.sqrt(varianza_datos1):.2f}")
print(f"Cuartiles (Q1,Q2,Q3): {cuartiles_datos1}")
print(f"Rango intercuartil:   {cuartiles_datos1[2]-cuartiles_datos1[0]:.2f}")
```

```python
# Resultados para el segundo conjunto de datos
print("\n" + "="*50)
print("ALGORITMO B (datos2)")
print("="*50)
print(f"Media:                {media_datos2:.2f}")
print(f"Mediana:              {mediana_datos2:.2f}")
print(f"Varianza:             {varianza_datos2:.2f}")
print(f"Desviacion estandar:  {np.sqrt(varianza_datos2):.2f}")
print(f"Cuartiles (Q1,Q2,Q3): {cuartiles_datos2}")
print(f"Rango intercuartil:   {cuartiles_datos2[2]-cuartiles_datos2[0]:.2f}")
```

# Diagrama de cajas
En la siguiente celda se muestra el código para generar un diagrama de cajas con los datos1 y datos2.

```python
# Preparar datos para boxplot
datos = [datos1, datos2]
etiquetas = ["Algoritmo A", "Algoritmo B"]

# Crear figura con tamano especifico
plt.figure(figsize=(8, 6))

# Generar diagrama de cajas
# patch_artist=True permite personalizar colores
plt.boxplot(datos,
            patch_artist=True,
            labels=etiquetas,
            boxprops=dict(facecolor='white', color='black'),
            whiskerprops=dict(color='black'),
            medianprops=dict(color='black'),
            capprops=dict(color='black'))

# Etiquetas y titulo
plt.ylabel('Costo de la solucion')
plt.title('Comparacion de algoritmos mediante diagrama de cajas')
plt.grid(True, alpha=0.3)
plt.show()
```
