# Funciones de prueba
Las funciones de prueba para técnicas heurísticas son, como su nombre lo sugiere, funciones o reglas de correspondencia empleadas en la evaluación y comparación del desempeño de algoritmos basados en métodos heurísticos.

Son diseñadas para incluir diferentes características que podrían dificultar la localización del óptimo, por ejemplo, existen funciones unimodales, multimodales, escalables, continuas, diferenciables, convexas, etc.

A continuación se presentan algunas funciones empleadas con frecuencia, formulando todos los problemas para minimización. Se debe mencionar que la propuesta de funciones de prueba es una línea de investigación en continuo desarrollo, por lo cual la presentación de una lista exhaustiva se encuentra fuera del alcance de este libro.

A partir de ahora se supondrá que $\overline{x}= x_1,x_2,...,x_n$.

Primero importamos la biblioteca NumPy que será empleada para crear una malla de puntos en dos dimensiones sobre la cual evaluar las funciones, y las bibliotecas matplotlib y mpl_toolkits para poder graficarlas en tres dimensiones.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

#Funciones unimodales
Las funciones unimodales se distinguen por tener un único óptimo. En esta sección se presenta la definición de tres ejemplos comunes: la función esfera, la función Rosenbrock y la función Easom.

Las dos primeras poseen la capacidad de extender su dominio a cualquier dimensión, un atributo conocido como escalabilidad. Por otro lado, la función Easom está definida exclusivamente para dos dimensiones.

##Función esfera
En el contexto de optimización, esta función también es conocida como la primera de De Jong. Es una función continua, convexa, y unimodal dada por:

```python
def esfera(x, y):
    return x**2 + y**2
```

Para crear una gráfica en tres dimensiones se empleará el código incluido en la siguiente celda, mismo que se repite con las funciones presentadas más adelante.

Los pasos realizados son los siguientes:

1. Crear una malla de puntos para graficar la función:
   - Se define un rango de valores para las coordenadas $x$ y $y$ utilizando `np.linspace`. En este caso, se generan 500 puntos uniformemente espaciados entre -5 y 5 tanto para $x$ como para $y$.
   - Se utiliza `np.meshgrid` para generar una cuadrícula bidimensional de puntos a partir de todas las combinaciones posibles de los valores de $x$ y $y$, con esto se obtienen las matrices X y Y.
   - Se calcula Z utilizando una función `esfera(X, Y)` que debería devolver los valores de Z correspondientes a cada par de coordenadas X y Y.

2. Graficar la superficie de la función:
   - Se utiliza `ax.plot_surface` para graficar la superficie tridimensional definida por las coordenadas X, Y y Z.

3. Etiquetas de los ejes:
   - Se utilizan `ax.set_xlabel`, `ax.set_ylabel` y `ax.set_zlabel` para etiquetar los ejes X, Y y Z, respectivamente.

5. Título:
   - Se utiliza `plt.title` para establecer un título para la figura.

6. Mostrar la figura:
   - Finalmente, se muestra la figura utilizando `plt.show()`.

Se debe destacar que se utiliza la misma secuencia de pasos para generar las gráficas siguientes, sólo se cambia la función empleada para obtener los valores de Z.


```python
# Crear una malla de puntos para graficar la función
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = esfera(X, Y)

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie de la función
ax.plot_surface(X, Y, Z, cmap='viridis')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Título
plt.title('Función esfera')

# Mostrar la figura
plt.show()
```

```python
# Crear una malla de puntos para graficar las curvas de nivel
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
X, Y = np.meshgrid(x, y)
Z = esfera(X, Y)

# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# Graficar las curvas de nivel con un degradado continuo
contour = plt.contourf(X, Y, Z, levels=30, cmap='viridis')

# Graficar las curvas de nivel con líneas negras para resaltarlas
plt.contour(X, Y, Z, levels=20, colors='black')

# Etiquetas de los ejes
plt.xlabel('X')
plt.ylabel('Y')

# Título
plt.title('Curvas de nivel de la función esfera')

# Agregar la barra de color para el valor Z
plt.colorbar(contour, label='Valor de Z')

# Colocar un círculo rojo en el mínimo de la función (en (0,0))
plt.plot(0, 0, 'ro', markersize=8)  # 'ro' es para un círculo rojo

# Mostrar la gráfica
plt.show()
```

##Función Rosenbrock
También conocida como el valle de Rosenbrock es una función no lineal que se caracteriza por tener el mínimo global dentro de un valle largo y angosto, por lo que resulta difícil de encontrar, está dada por:

```python
def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2
```

En esta ocasión se incluye `ax.view_init(elev=15, azim=45)` para ajustar los ángulos de visión del gráfico. `elev=15` especifica una elevación de 15 grados, y `azim=45` especifica un azimut de 45 grados. Puedes modificarlos para que veas cómo se rota la imágen.

```python
# Crear una malla de puntos para graficar la función
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie de la función
ax.plot_surface(X, Y, Z, cmap='viridis')

# Ajustar ángulos de visión
ax.view_init(elev=15, azim=45)  # Elevación de 30 grados, azimut de 45 grados

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Título
plt.title('Función Rosenbrock')

# Mostrar la figura
plt.show()
```

```python
# Crear una malla de puntos para graficar las curvas de nivel
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
X, Y = np.meshgrid(x, y)
Z = rosenbrock(X, Y)

# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# Graficar las curvas de nivel con un degradado continuo
contour = plt.contourf(X, Y, Z, levels=30, cmap='viridis')

# Graficar las curvas de nivel con líneas negras para resaltarlas
plt.contour(X, Y, Z, levels=20, colors='black')

# Etiquetas de los ejes
plt.xlabel('X')
plt.ylabel('Y')

# Título
plt.title('Curvas de nivel de la función de Rosenbrock')

# Agregar la barra de color para el valor Z
plt.colorbar(contour, label='Valor de Z')

# Colocar un círculo rojo en el mínimo de la función (en (1,1))
plt.plot(1, 1, 'ro', markersize=8)  # 'ro' es para un círculo rojo

# Mostrar la gráfica
plt.show()
```

## Función Easom
La función de Easom se caracteriza por tener muchos mínimos locales y un solo mínimo global ubicado en una región muy pequeña en comparación con el resto del dominio, está dada por:

```python
def easom(x, y):
    return -np.cos(x) * np.cos(y) * np.exp(-((x - np.pi)**2 + (y - np.pi)**2))
```

La gráfica se obtiene empleando el mismo código que con la función esfera.

```python
# Crear una malla de puntos para graficar la función
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = easom(X, Y)

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie de la función
ax.plot_surface(X, Y, Z, cmap='viridis')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Título
plt.title('Función Easom')

# Mostrar la figura
plt.show()
```

```python
# Crear una malla de puntos para graficar las curvas de nivel
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = easom(X, Y)

# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# Graficar las curvas de nivel con un degradado continuo
contour = plt.contourf(X, Y, Z, levels=30, cmap='viridis')

# Graficar las curvas de nivel con líneas negras para resaltarlas
plt.contour(X, Y, Z, levels=20, colors='black')

# Etiquetas de los ejes
plt.xlabel('X')
plt.ylabel('Y')

# Título
plt.title('Curvas de nivel de la función de Easom')

# Agregar la barra de color para el valor Z
plt.colorbar(contour, label='Valor de Z')

# Colocar un círculo rojo en el mínimo de la función (en (pi, pi))
plt.plot(np.pi, np.pi, 'ro', markersize=8)  # 'ro' es para un círculo rojo

# Mostrar la gráfica
plt.show()
```

#Funciones multimodales
Las funciones multimodales son aquellas que tienen múltiples óptimos, locales o globales. En esta sección se incluye la definición de las funciones de Ackley, Griewank, Rastrigin y Alpine. Las cuatro funciones son escalables en el dominio y su contradominio son los números reales.

Observa que en todos los casos el rango de las funciones fue restringido a $[-5,5]\times[-5,5]$ para facilitar la visualización de los óptimos locales. Sin embargo, si deseas emplear estas funciones para comparar el desempeño de diferentes técnicas heurísticas es recomendable usar los rangos dados en la definición original de cada función, como se muestra en el libro.

##Función de Ackley
La función de Ackley está dada por:

```python
def ackley(x, y):
    a = 20
    b = 0.2
    c = 2 * np.pi
    term1 = -a * np.exp(-b * np.sqrt((x**2 + y**2) / 2))
    term2 = -np.exp((np.cos(c * x) + np.cos(c * y)) / 2)
    return term1 + term2 + a + np.exp(1)
```

Ahora se puede realizar la gráfica.

```python
# Crear una malla de puntos para graficar la función
x = np.linspace(-5, 5, 500)
y = np.linspace(-5, 5, 500)
X, Y = np.meshgrid(x, y)
Z = ackley(X, Y)

# Crear una figura 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Graficar la superficie de la función
ax.plot_surface(X, Y, Z, cmap='viridis')

# Etiquetas de los ejes
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Título
plt.title('Función de Ackley')

# Mostrar la figura
plt.show()
```

```python
# Crear una malla de puntos para graficar las curvas de nivel
x = np.linspace(-5.0, 5.0, 100)
y = np.linspace(-5.0, 5.0, 100)
X, Y = np.meshgrid(x, y)
Z = ackley(X, Y)

# Configurar el tamaño de la figura
plt.figure(figsize=(8, 6))

# Graficar las curvas de nivel con un degradado continuo
contour = plt.contourf(X, Y, Z, levels=30, cmap='viridis')

# Graficar las curvas de nivel con líneas negras para resaltarlas
plt.contour(X, Y, Z, levels=20, colors='black')

# Etiquetas de los ejes
plt.xlabel('X')
plt.ylabel('Y')

# Título
plt.title('Curvas de nivel de la función de Ackley')

# Agregar la barra de color para el valor Z
plt.colorbar(contour, label='Valor de Z')

# Colocar un círculo rojo en el mínimo global (en (0, 0))
plt.plot(0, 0, 'ro', markersize=10)  # Círculo rojo para el mínimo global

# Coordenadas de los óptimos locales encontrados
local_minima = [
    (0.96847765, 0.96847765),
    (-0.96847766, 0.96847765),
    (0.96847765, -0.96847766),
    (-0.96847766, -0.96847766),
    (0, 0.952166542),
    (0, -0.952166531),
    (0.952166542, 0),
    (-0.952166531, 0)
]

# Graficar los óptimos locales con círculos de tamaño más pequeño y color contrastante (amarillo)
for (lx, ly) in local_minima:
    plt.plot(lx, ly, 'orange', marker='o', markersize=6) # Círculos amarillos para óptimos locales

# Mostrar la gráfica
plt.show()
```

## Función Griewank
La función de Griewank es una función multimodal con muchos mínimos locales distribuidos de manera regular, dada por:

```python
def griewank(x, y):
    return 1 + (x**2 / 4000) + (y**2 / 4000) - np.cos(x) * np.cos(y / np.sqrt(2))
```

##Función de Rastrigin

##Función de Alpine
La función de Alpine es una función no convexa y multimodal, dada por:

```python
def alpine(x, y):
    return np.abs(x * np.sin(x) + 0.1 * x) + np.abs(y * np.sin(y) + 0.1 * y)
```
