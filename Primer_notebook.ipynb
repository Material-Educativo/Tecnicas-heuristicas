# Condicionales $if$ y $else$

Al igual que en muchos otros lenguajes de programación, los condicionales en Python incluyen las funciones *if*, y *else*, que se emplean para realizar un conjunto de acciones cuando se satisface alguna condición. En la siguiente celda de código se muestra cómo puedes emplearlas.


```python
# Definir una variable
temperatura = 28

# Evaluar la condición
if temperatura > 25:
    print("Hace calor, usa ropa ligera")
else:
    print("Hace frío, usa suéter")
```

## Condicionales múltiples con $elif$.

```python
# Calificación según puntuación
puntos = 85

if puntos >= 90:
    print("Calificación: A (Excelente)")
elif puntos >= 80:
    print("Calificación: B (Bien)")
elif puntos >= 70:
    print("Calificación: C (Suficiente)")
else:
    print("Calificación: F (Reprobado)")
```

# Ciclo *for*
En Python se puede usar el *for* de diferentes maneras, pero la forma más sencilla consiste en definir el intervalo de valores que puede tomar la variable que sirve como contador del ciclo. En la siguiente celda se ha empleado la variable *i* como contador en el ciclo, y se ha escrito *in range(10)* para indicar que *i* debe tomar valores enteros del 0 hasta el 9.

El resultado al ejecutar esta celda será la impresión de los valores que va tomando la variable *i* en cada paso del ciclo, empezando desde 0 y terminando en 9.

```python
# Imprimir números del 0 al 9
for i in range(10):
    print("Iteración número:", i)

# El ciclo automáticamente incrementa i de 0 a 9
```

Ejemplo con una tabla de multiplicar.

```python
# Tabla del 7
numero = 7

print(f"Tabla de multiplicar del {numero}:")
print("-" * 20)

for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

# Ciclo *while*

El uso del ciclo *while* se ilustra en la siguiente celda. Observa que las instrucciones de las líneas 3 y 4 están desplazadas hacia la derecha con respecto al margen de la línea 2. Por lo tanto, el valor de la variable *i* se imprimirá diez veces, mostrando su aumento de uno en uno.

```python
# Contar hasta 10
contador = 1

while contador <= 10:
    print("Conteo:", contador)
    contador += 1  # Equivale a: contador = contador + 1

print("Terminé de contar!")
```

Para dar otro ejemplo de la importancia del sangrado en Python se modificará el código anterior borrando el sangrado de la línea 4, como se muestra en la siguiente celda.

En esta versión la función *print* ya no está dentro del ciclo *while*, así que la impresión de la variable *i* sólo se realizará una vez, cuando se salga del ciclo.

```python
# Ejemplo 1: print() DENTRO del ciclo
i = 0
while i < 3:
    i += 1
    print("Dentro del ciclo:", i)  # Se imprime 3 veces

# Ejemplo 2: print() FUERA del ciclo
j = 0
while j < 3:
    j += 1
print("Fuera del ciclo:", j)  # Se imprime 1 vez (al final)
```

# Definición de funciones
En algunas ocasiones es necesario definir funciones propias del problema que se está analizando. Para lo cual se requiere crear funciones que reciban parámetros, realicen algunas operaciones con ellos, y que devuelvan los resultados.

Primero se verá una función que recibe dos números, los suma y devuelve el resultado

```python
def nombre_funcion(parametro1, parametro2):
    """
    Docstring: Breve descripción de qué hace la función

    Parámetros:
    parametro1 -- descripción
    parametro2 -- descripción

    Retorna:
    descripción de lo que devuelve
    """
    # Código de la función (con 4 espacios de indentación)
    resultado = parametro1 + parametro2

    # Devolver el resultado
    return resultado
```

## Creamos una función.

```python
def suma(a, b):
    """
    Suma dos números y devuelve el resultado.

    Parámetros:
    a -- primer número
    b -- segundo número

    Retorna:
    La suma de a + b
    """
    resultado = a + b
    return resultado
```

Una vez que has creado una función debes compilar la celda, con el ícono de *play* o con la combinación de teclas *Shift + Enter*.

Recuerda compilar las celdas de las dos funciones creadas hasta este punto.

Para poder usar las funciones es necesario emplear su nombre completo, proporcionar los parámetros que requieren de acuerdo a su definición, y establecer las variables adecuadas para recibir los valores que devuelvan.

En la siguiente celda se ejemplifica cómo llamar a la función *Suma*.

El valor devuelto por la función *Suma* se guarda en la variable *total*, o en la variable $mi\_suma$, y se imprime.

```python
# ===== USAR LA FUNCIÓN =====

# Llamar a la función con valores específicos
total = suma(5, 3)
print("5 + 3 =", total)  # Imprime: 5 + 3 = 8

# Llamar con variables
x = 10
y = 20
mi_suma = suma(x, y)
print(f"{x} + {y} = {mi_suma}")  # Imprime: 10 + 20 = 30
```

# Funciones con múltiples retornos

```python
def suma_resta(a, b):
    """
    Realiza suma y resta de dos números.

    Parámetros:
    a -- primer número
    b -- segundo número

    Retorna:
    Una tupla con (suma, resta)
    """
    suma = a + b
    resta = a - b

    # Indicar si los resultados son negativos
    if suma < 0:
        print("La suma es negativa")

    if resta < 0:
        print("La resta es negativa")

    # Devolver ambos resultados
    return suma, resta
```

En este caso la función nos devuelve varios valores, y necesitamos una variable diferente para recibir cada uno de ellos. En la siguiente celda usamos las variables $resultado_suma$ y $resultado_resta$ para guardar e imprimir los valores devueltos por la función $suma\_resta$.

```python
# ===== USAR LA FUNCIÓN =====

# Capturar ambos valores devueltos
resultado_suma, resultado_resta = suma_resta(10, 7)

print(f"Suma: {resultado_suma}")    # Suma: 17
print(f"Resta: {resultado_resta}")  # Resta: 3
```

# Funciones sin retorno.

```python
def saludar(nombre):
    """
    Imprime un saludo personalizado.
    No devuelve ningún valor.
    """
    print(f"Hola, {nombre}! Bienvenido al curso de heurísticas.")
```

```python
# Usar la función
saludar("Ana")     # Imprime: Hola, Ana! Bienvenido...
saludar("Carlos")  # Imprime: Hola, Carlos! Bienvenido...

# Si intentas guardar el resultado:
resultado = saludar("Pedro")
print(resultado)  # Imprime: None (porque no hay return)
```
