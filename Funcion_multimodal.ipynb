```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcion_multimodal(x):
    return x**2 / 10 + np.sin(5 * x)  # más oscilaciones por el factor 5
```

```python
x_vals = np.linspace(-5, 5, 1000)
y_vals = funcion_multimodal(x_vals)

plt.figure(figsize=(12, 6))
plt.plot(x_vals, y_vals, 'b-', linewidth=2, label='f(x)')
plt.xlabel('x', fontsize=12)
plt.ylabel('f(x)', fontsize=12)
plt.title('Función multimodal con valles pronunciados', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()
```
